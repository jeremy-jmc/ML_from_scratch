```python
import numpy as np      # linear algebra
import matplotlib.pyplot as plt     # plot
import pandas as pd         # datasets
import seaborn as sns       # beautiful plots
import warnings
import random
warnings.filterwarnings('ignore')
```

```python
def gradient_descent(start, gradient, learn_rate, max_iter, tol=0.01):
    """Gradient descent algorithm.
    
    args:
        start: starting point (maybe random initialisation)
        gradient: function to calculate gradient
        learn_rate: learning rate | scaling factor for step sizes
        max_iter: maximum number of iterations
        tol: tolerance for stopping criterion

    returns:
        steps: list of all steps
        x: final value
    """
    steps = [start]
    x = start

    for _ in range(max_iter):
        diff = learn_rate * gradient(x)
        if np.abs(diff) < tol:
            break    
        x = x - diff
        steps.append(x)
    return steps, x
```

```python
def func1(x):
  return x**2 - 4*x + 1

def gradient_func1(x):
  return 2*x - 4
```

```python
# 1. Run GD algorithm
alpha = 0.1
history, result = gradient_descent(
    start = -5,   # random.randint(-6, 10)
    gradient = gradient_func1, 
    learn_rate = alpha, 
    max_iter = 200
)

# 2. Plot the results
x = np.linspace(-6, 10, 100)    # 100 evenly spaced numbers between -6 and 10
y = func1(x)

fig = plt.figure(figsize=(8, 6))
sns.lineplot(x, y)

print(len(history))

steps = history

for i in range(0, len(steps)):
    sns.lineplot(steps[i:i+2], func1(np.array(steps[i:i+2])), color='red', marker='o')
fig.suptitle(f'Learning rate: {alpha}')
```

```python
func2 = lambda x: x**4 - 2*(x**3) + 2
gradient_func2 = lambda x: 4*(x**3) - 6*(x**2)

# 1. Run GD algorithm
alpha = 0.09
history, result = gradient_descent(
    start = -0.5,   # random.randint(-6, 10)
    gradient = gradient_func2, 
    learn_rate = alpha, 
    max_iter = 500      # prevent divergence
)   # ! WARNING: si se muere el alg es por la precision


# 2. Plot the results
x = np.linspace(-0.5, 2, 100)    # 100 evenly spaced numbers between -6 and 10
y = func2(x)

fig = plt.figure(figsize=(8, 6))
sns.lineplot(x, y)

steps = history
for i in range(0, len(steps)):
    sns.lineplot(steps[i:i+2], func2(np.array(steps[i:i+2])), color='red', marker='o')
fig.suptitle(f'Learning rate: {alpha}')
```
